# Part 1 - Creating a Function

```python
# Basic syntax
# def function_name():
# code
```

```python
def greet(): # function with no parameters
    print("Hello") # it does not return anything, you must call the function to see the output 
greet() # calling the function to see the output
```

# Part 2 - function with Parameter

```python
def greet(name): # function with one parameter, you can pass any name as an argument to the function
    print(f"Hello {name}")
greet("kunal")    
```

```python
def square(number):
    print(number * number)

square(23)
```

# Part 3 - Multiple Parameters

```python
def add(a, b): # function with two parameters, you can pass any two numbers as arguments to the function
    print(a + b)

add(10, 20)
```

```python
# Real Analytics Example
def calculate_revenue(price, quantity):
    print(price * quantity)

calculate_revenue(500, 10)
```

# Part 4 - Return Statement

```python
def add(a, b):
    print(a + b) # without return statement, the function will return None by default
```

```python
def add(a, b):
    return a + b
result = add(10, 20)
print(result) # with return statement, the function will return the value and you can store it in a variable and use it later
```

```python
def multiply(a, b):
    return a * b

result = multiply(5, 4)

print('the result is:', result)
```

# Part 5 - Default Parameters

```python
def greet(name="Guest"):
    print(f"Hello {name}")
greet() # calling the function without passing any argument, it will use the default value "Guest"
greet("Kunal") # calling the function with an argument, it will override the default value and use "Kunal" instead
```

# Part 6 - Local vs Global Variables

```python
# Local variables
def test():  # exists only inside the function, you cannot access it outside the function
    x = 10
    print(x)

test()
```

```python
# Global variables
name = "Kunal"  # exists outside the function, you can access it inside the function

def greet():
    print(name)

greet()
```

# Part 7 - functions calling other functions

```python
def greet():
    print("Hello")

def welcome():
    greet()
    print("Welcome")

welcome()
```

# Part 8 - Why functions matter in Data Analysis
 Functions help:

        Reuse KPI calculations
        Build automation scripts
        Create reusable analytics logic
        Simplify notebooks

```python
# Example - Profit Margin Function
def profit_margin(revenue, profit):

    margin = (profit / revenue) * 100

    return margin
result = profit_margin(100000, 25000)

print(result)
```

# Part 9 - scope of Functions
Functions isolate logic

```python
def calculate_tax(amount):

    tax = amount * 0.18

    return tax
result = calculate_tax(50000)
print(result)
```
