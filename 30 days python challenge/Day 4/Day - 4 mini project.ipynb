# Day 4 Mini Project - Business Kpi Calculator
This project combines:

    Functions
    Parameters
    Return values
    Business calculations

```python
# BUSINESS KPI CALCULATOR

def revenue(price, quantity):

    return price * quantity


def profit(revenue, cost):

    return revenue - cost


def profit_margin(profit, revenue):

    return (profit / revenue) * 100


# INPUTS

price = float(input("Enter product price: "))

quantity = int(input("Enter quantity sold: "))

cost = float(input("Enter total cost: "))


# CALCULATIONS

total_revenue = revenue(price, quantity)

total_profit = profit(total_revenue, cost)

margin = profit_margin(total_profit, total_revenue)


# OUTPUT

print("\n===== KPI REPORT =====")

print(f"Revenue: ₹{total_revenue}")

print(f"Profit: ₹{total_profit}")

print(f"Profit Margin: {margin:.2f}%")
```
