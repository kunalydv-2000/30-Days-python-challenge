# Task 1- Area Calculator 

```python
# Area calculator
def area(length, width):

    return length * width


result = area(10, 5)

print(result)
```

# Task 2 - Employee Bonus Calculator

```python
# Employee Bonus calculator
def bonus(salary):

    return salary * 0.10


print(bonus(50000))
```

# Task 3 - Even/odd function

```python
# Even/odd function
def even_odd(num):

    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
print(even_odd(10))
print(even_odd(7))  
```
