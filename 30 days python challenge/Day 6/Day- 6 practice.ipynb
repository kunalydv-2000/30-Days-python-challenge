```python
employee = {
    "name": "Rahul",
    "salary": 50000
}

employee["salary"] = 70000 # update

print(employee)
```

# Part 7- Looping through dictionaries

```python
# loop through keys
student = {
    "name": "Kunal",
    "age": 21
}

for key in student:
    print(key)
```

```python
# loop through values
for value in student.values():
    print(value)
```

```python
# loop through both key and value
for key, value in student.items():
    print(key,'-',value)
```

# Part 10- What is a **Set** ?
A set stores:

    -Unique values only
    -Unordered data

```python
numbers = {1, 2, 3, 4,5,1,2,3,12,45,6,2,78,21,2,45,3}   # sets removes duplicates automatically
numbers
```

```python
data = {1, 2, 2, 3, 3, 4}

print(data)
```

# why sets matter in data analysis?

# Part 12- Adding and Removing set elements

```python
# add
numbers = {1, 2, 3}

numbers.add(4)

print(numbers)
```

# Part 13- Set Operations

```python
# union - combines sets
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)
```

```python
# intersection - common values
a = {1, 2, 3}
b = {3, 4, 5}

print(a & b)
```

```python
# difference - values in first set only
print(a - b)
```

# Real Data Analyst Uses
    Dictionaries are heavily used in:

JSON APIs
1.Configuration files

2.KPI mappings

3.Data transformation

4.Frequency counting

    Sets are used in:

1.Duplicate removal

2.Unique value analysis

3.Customer overlap analysis

```python
# frequency counter
sales_regions = [
    "North",
    "South",
    "North",
    "East"
]

count = {}

for region in sales_regions:

    if region in count:
        count[region] += 1

    else:
        count[region] = 1

print(count)
```
