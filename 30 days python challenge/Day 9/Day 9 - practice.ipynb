# Part 2- What is a Series?
A series is:
- one-dimensional
- like a single column

---

```python
# Example 
import pandas as pd

sales = pd.Series([1000, 2000, 3000])

print(sales)
```

# Pat 3- Creating a DataFrame
A DataFrame is a table.

---

```python
# Example
import pandas as pd

data = {

    "Name": ["Kunal", "Rahul", "Aman"],

    "Sales": [5000, 7000, 6000],

    "Region": ["North", "South", "East"]

}

df = pd.DataFrame(data)

print(df)
```

# Part 4- Reading CSV files

```python
import csv

with open("sales.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerow(["Name", "Sales", "Region"])

    writer.writerow(["Kunal", 5000, "North"])

    writer.writerow(["Rahul", 7000, "South"])
```

```python
import pandas as pd

df = pd.read_csv("sales.csv")

print(df)
```
