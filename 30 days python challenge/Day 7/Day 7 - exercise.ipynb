# Task 3 - Student CSV Writer
---

```python
# Example
import csv

with open("students.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerow(["Name", "Marks"])

    writer.writerow(["Kunal", 85])

    writer.writerow(["Rahul", 90])
```

```python
import csv

with open("students.csv", "r") as file:

    reader = csv.reader(file)

    next(reader)

    for row in reader:

        Name = row[0]

        Marks = row[1]

        print(Name, Marks)
```

```python
import csv

total_marks = 0

with open("students.csv", "r") as file:

    reader = csv.reader(file)

    next(reader)

    for row in reader:

        total_marks += int(row[1])

print(total_marks)
```
