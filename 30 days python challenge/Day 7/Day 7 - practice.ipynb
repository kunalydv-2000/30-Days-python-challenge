# Part - 3 -Reading Files
---

Example File (`AUDIENCE DATABASE.sql`)

````python
file = open('AUDIENCE DATABASE.sql','r')

content = file.read()
print(content)
file.close()
````

# Important - Close files
always close files.

`file.close()`

this:
- Saves memory 
- prevents corruption

# Part 4 - `with open()` (Best Practice)
---
Better Approach


````python
## Example 
with open("AUDIENCE DATABASE.sql", "r") as file:

    content = file.read()

    print(content)
````

## Why Better?
Python automatically closes files.

Recommended in professional coding.


# Part 5 - Reading Line By Line
---

````python
# Example 
with open("AUDIENCE DATABASE.sql", "r") as file:

    for line in file:   # for large dataset where reading full file is inefficient
        print(line)
````

# Part 6 - Writing Files
---
write mode creates or overwrite files


````python
# Example
with open("report.txt", "w") as file:

    file.write("Sales Report")
````

# Important warning
'w' mode deletes old content before writing.

# Part 7 - Appending Files
---
Appends add data without deleting existing content. 

````python
# Example
with open("report.txt", "a") as file:

    file.write("\nNew Sales Data")
````

## Understanding `\n`
Creates new line

# Part 11 - Writing Csv Files
-----

````python
# Example
import csv

with open("sales.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerow(["Name", "Sales"])

    writer.writerow(["Kunal", 5000])

    writer.writerow(["Rahul", 7000])
````

````python
## Example 
with open("sales.csv", "r") as file:

    content = file.read()

    print(content)
````

# Part 12 - Appending CSV Rows
---

````python
import csv

with open("sales.csv", "a", newline="") as file:

    writer = csv.writer(file)

    writer.writerow(["Aman", 9000])
````

````python
# Example 
with open("sales.csv", "r") as file:

    content = file.read()

    print(content)
````

# Part 13 - Reading CSV into Variables
---

````python
# Example
import csv

with open("sales.csv", "r") as file:

    reader = csv.reader(file)

    next(reader) # skips header

    for row in reader:

        name = row[0]

        sales = row[1]

        print(name, sales)
````

# Part 14 - CSV Calculations
---

````python
# Example
import csv

total_sales = 0

with open("sales.csv", "r") as file:

    reader = csv.reader(file)

    next(reader)

    for row in reader:

        total_sales += int(row[1])

print('total sales =',total_sales)
````
