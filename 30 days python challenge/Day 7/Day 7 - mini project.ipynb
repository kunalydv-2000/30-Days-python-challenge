# Day 7 Mini Project - CSV Sales Logger
This projects combines:
- File handling 
- CSV writing
- CSV reading
- Loops 
- Calculations
--- 

```python
import csv

while True:

    print("\n===== SALES LOGGER =====")

    print("1. Add Sale")

    print("2. View Sales")

    print("3. Total Sales")

    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        name = input("Enter product name: ")

        amount = float(input("Enter sales amount: "))

        with open("sales_new.csv", "a", newline="") as file:

            writer = csv.writer(file)

            writer.writerow([name, amount])

        print("Sale Added Successfully")

    elif choice == "2":

        print("\n===== SALES RECORDS =====")

        with open("sales_new.csv", "r") as file:

            reader = csv.reader(file)

            for row in reader:
                print(row)

    elif choice == "3":

        total = 0

        with open("sales_new.csv", "r") as file:

            reader = csv.reader(file)

            for row in reader:

                total += float(row[1])

        print(f"Total Sales: ₹{total}")

    elif choice == "4":

        print("Exiting Program")

        break

    else:
        print("Invalid Choice")
```
