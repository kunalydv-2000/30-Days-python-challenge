# Task 1- Student Marks Analysis
---

```python
# Student Marks Analysis
import numpy as np

marks = np.array([85, 90, 78, 92, 88])

print('Mean of marks are',marks.mean())

print('Maximum marks are',marks.max())

print('Minimum marks are',marks.min())

print('Total marks are',marks.sum())
```

# Task 3- Even number filter

```python
import numpy as np

numbers = np.array([1, 2, 3, 4, 5, 6])

print('Even numbers are',numbers[numbers % 2 == 0])

print('Odd numbers are',numbers[numbers % 2 != 0])
```
