# Day 8 Mini Project — Sales Array Analysis

This project combines:

- Arrays
- Aggregations
- Analytics calculations

```python
# Sales Array Analysis
import numpy as np

sales = np.array([

    12000,
    15000,
    18000,
    20000,
    17000,
    22000

])

print("===== SALES ANALYSIS =====")

print(f"Sales Data: {sales}")

print(f"Total Sales: ₹{sales.sum()}")

print(f"Average Sales: ₹{sales.mean():.2f}")

print(f"Highest Sales: ₹{sales.max()}")

print(f"Lowest Sales: ₹{sales.min()}")

print(f"Standard Deviation: ₹{sales.std():.2f}")
```
