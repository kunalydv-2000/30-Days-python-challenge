# Part 2 - What is an array ?
---
An array is similar to a List.

BUT:
- faster
- more powerful
- optimized for numerical operations

````python
# Numpy Array Example
import numpy as np

numbers = np.array([1, 2, 3, 4])

print(numbers)
````

## Difference b/w List and Array
| Feature | List | NumPY Array |
| --- | --- | --- |
| Speed | Slower | Faster |
| Numerical operations | Limited | Advanced |
| Memory efficiency | Lower | Higher |
| Data Analysis use | Basic | Professional |

# Part 3 - Creating Arrays
---

````python
# Integer array
import numpy as np

arr = np.array([1, 2, 3, 4])

print(arr)
````

````python
# Float array
arr = np.array([1.5, 2.5, 3.5])

print(arr)
````

````python
# Mixed
arr = np.array([1, 2.5, 3])
print(arr)
````

# Part 4 - Array Attributes
---

````python
# Shape
arr = np.array([1, 2, 3])
                            # 3 elements,1-dimensional array
print(arr.shape)
````

````python
# data type 
print(arr.dtype)
````

````python
# Size
print(arr.size)
````

# Part 5 - Array Indexing
---
similar to list


````python
arr = np.array([10, 20, 30, 40])

print(arr[3])
````

# Part 6 - Array Slicing

````python
arr = np.array([10, 20, 30, 40, 50])

print(arr[2:])    #include : start index  and exclude : end index
````

# Part 7 - Array Operations
---
Unlike lists, array allow direct mathematical operations 

````python
# Example - Addition
arr = np.array([1, 2, 3])

print(arr + 5) # numpy adds 5 to every element
````

````python
# Example - Multiplication
arr = np.array([2,4,6])

print(arr * 3)
````

# Part 8 - Array Arithmetic 
---

````python
# Addition b/t arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
````

````python
# Multiplication
print(a * b)
````

# Part 9-Aggregation Functions
---

````python
# Sum
sales = np.array([1000, 2000, 3000])

print(sales.sum())
````

````python
# Mean
sales = np.array([1000, 2000, 3000])

print(sales.mean())
````

````python
# max 
sales = np.array([1000, 2000, 3000])

print(sales.max())
````

````python
# min
sales = np.array([1000, 2000, 3000])

print(sales.min())
````

````python
# standard deviation
sales = np.array([1000, 2000, 3000])

print(sales.std())
````

# Part 10 - Creating special arrays
---

````python
# zero arrays
arr = np.zeros(5)

print(arr)
````

````python
# ones array
arr = np.ones(5)

print(arr)
````

````python
# range array
arr = np.arange(1, 11, 2)

print(arr)
````

# Part 11 - Multi - Dimensional Arrays
---
NumPY supports table/matrices 

````python
# 2 - D Array example
data = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(data)
````

````python
# shape
print(data.shape)
````

Meaning:
- 2 rows
- 3 columns

# Part 12 - Reshaping Arrays
---

````python
# Example
arr = np.arange(1, 7)

new_arr = arr.reshape(2, 3)

print(new_arr)
````

## Understanding
converted:
- 1-D array
    → into
- 2-D array
