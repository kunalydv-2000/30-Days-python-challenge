# Part 8 - Loop through Lists

```python
sales = [1000, 2000, 3000]

for amount in sales:
    print(amount)
```

```python
# real Analytics example
sales = [1000, 2000, 3000]

total = 0

for amount in sales:
    total += amount

print(total)
```

# Part 9 - list slicing
 slicing extracts portions of a list

```python
# list[start:end]
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4]) # includes a start and end point
```

```python
numbers = [10, 20, 30, 40, 50]

print(numbers[:4]) # only end point
```

```python
numbers = [10, 20, 30, 40, 50]

print(numbers[1:]) # only start point
```

# Part 11 - Nested lists
list inside list

```python
data = [
    ["Kunal", 50000],
    ["Rahul", 60000],
    ["Aman", 55000]
]
print(data[2])
```

```python
data = [
    ["Kunal", 50000],
    ["Rahul", 60000],
    ["Aman", 55000]
]
print(data[2][1]) # indexing in nested list
```

# Part 12 - **Tuples**

```python
# difference b/w list and tuples
#feature        list       tuple
#syntax         []         ()
#mutable        yes        no
#changeable     yes        no
```
